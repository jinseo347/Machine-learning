```python
# 두번째꺼. 
# spade a = 3/4, heart K = 1/4
p_a = 3/4
a=-p_a*np.log2(p_a)
p_k = 1/4
k=-p_k*np.log2(p_k)
H2=a+k
H2
```

```python
#네번째꺼
# spade a = 1/2, heart K = 1/2, diamond 5 = 1/4
p_a = 1/2
a=-p_a*np.log2(p_a)
p_k = 1/4
k=-p_k*np.log2(p_k)
p_d = 1/4
d=-p_d*np.log2(p_d)
H4=a+k+d
H4
```
