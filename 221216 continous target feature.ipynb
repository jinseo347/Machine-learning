#### underfitting과 overfitting
- Underfitting: 훈련이 아직 덜 된 경우.
- Overfitting: 모델이 너무 복잡해진 경우. depth가 너무 깊어져서 분기가 너무 많이 일어남. 
    - 이를 방지하기 위한 pruning이라는 가지치기 방법이 있음. 
        - pre-pruning: Decition tree 그리기 전 미리 제약사항을 만듦.
            - Early stoping creation: usually around 5%. Depth 설정.
        - post-pruning: Decision tree가 완성되고나서 하나 하나 확인. 
    - Decision tree는 Overfitting의 가능성을 가지고 있음. 
        - 그러나 설명은 주로 polynominal regression으로 설명. 
        
        https://www.microsoft.com/en-us/research/people/cmbishop/prml-book/

            - 3차시에서는 정답은 2개만 지나가지만 원하는 형태와 비슷. 
            - 9차시의 데이터와 3차시의 데이터 비교해보면 mean squared error 커짐.
