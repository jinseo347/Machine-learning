### Entropy(무질서도)

- Impurity metrics 중에 하나
- 얼마나 불순물이 섞여있는지, 얼마나 무질서한지를 나타내는 지표.
- uncertainty


#### entropy의 식.

- 확률P: 어떤 사건이 일어날 가능성.
- Sigma i부터 l까지(descriptive feature의 갯수?)
- 임의의 샘플을 뽑았을 때, i라는 descriptive feature의 갯수.
- 구한 확률에 로그. 
- 두개를 곱하고 더하면 data set의 엔트로피가 됨.
- '-'값인 이유, 로그 일반식을 보았을때, 0~1(확률)의 값은 음수 값이므로
- log의 밑이 2인 이유: computer는 0과 1이 정보의 단위(bit)
- 정보에 따라 불확실성이 감소하는 방향으로 decision tree가 형성. 
- 정보: degree of surprisal/ 정보의 가중치가 다름. 

참고. 로그 계산 주피터 노트북으로 하기 
```python
import numpy as np
p = 1/3
np.log2(p)
```

엔트로피의 계산

- 클래스가 두개일때 같은 갯수로 존재하면 그때의 엔트로피는 1!
- 두개의 값(binary class)으로 분류될때 엔트로피의 최댓값은 1: 어떻게 해도 분류할수 없음.  
- decision tree의 목표는 클래스를 분류하는 것
    - 확인 가능한 descriptive feature가 존재하지 않거나, 노이즈가 존재하는 경우에는 다수결(Majority)로 자료를 분류. 
    - leaf node 안에 가장 많은 분류로 classfy.
    - 반반일 경우 어느 쪽으로도 classify 분류 불가.

Weight 가중치
- 더 중요한 것에 더 큰 가중치를 둔다.

- 아랫첨자 d: 특정 descriptive feature로 분기 후의 dataset.
- 절대값 안의 로테이션 의미: 데이터 집합
- 분모: Large D. 전체 데이터 셋, 집합의 절대값, 집합의 원소의 갯수
- 분자: small d. 분기 이후의 나뉜 후의 데이터 셋의 크기.
- 즉, 더 큰 데이터 셋일수록, 가중치가 커진다.

가중치 부여 이유
- 분류가 다 끝난 dataset의 경우, 노이즈가 포함된 dataset 존재 가능성 있음. 
- 만약 하나짜리 노이즈에 같은 가중치로 계산된다면 정확하지 않음. 
- 분기된 dataset이 클수록 조금더 깨끗하고 잘 분류되어 있다고 봄. 

### information gain
- 데이터가 분기됨에 따라 정보를 얻고, 불확실성이 낮아짐. 
- 원래 dataset의 entropy와 하위 dataset의 entropy의 차이가 클수록 분류를 잘 했음을 의미. 
    - 엔트로피가 크게 줄어들음. 불확실성이 크게 줄어들음.
- descriptive feature를 잘 선택하는 기준이 될 수 있음. 

Information gain 계산하기



- 나눠지기 전의 엔트로피: 1
- suspicious words로 분기 되었을 때 
    - 각 엔트로피 0
    - 가중치 각 1/2
    - information gain: 1    
- unknown sender
    - 첫번째 entropy: 0.38997500048077083
    - 두번째 entropy: 0.5283208335737187
    - 가중치W: 각 1/2
    - information gain: 0.08170416594551044    
- Contains Images
    - 첫번째 entropy: 1
    - 두번째 entropy: 1
    - 가중치W: 각 1/3, 2/3
    - information gain: 0

````python
#information gain 계산하기

import numpy as np

#Unknown sender
s=2/3*np.log2(2/3)
h=1/3*np.log2(1/3)
p=-(s+h)

H = 0.5*p*2
IG2=1-H

# Contains Images
a=1*(1/3)
b=1*(4/6)
IG3=1-(a+b)

print(IG2)
print(IG3)
````

#### Review

- Decision tree: 목표- class의 분류, 효율적 분류를 위해 descriptive feature 중요.
- Entropy: 불확실성. descriptive feature 선택을 위해 계산 필요.
- Information Gain: 커질수록 불확실성이 감소됨. 즉 descriptive feature를 잘 선택했음을 의미.
