#### Entropy 공식 추가 설명

- 왜 '-'기호가 붙는가?
- 정보량 I = log2 (1/(p(x))=log2(P(x)-1)=-log2(P(x))

- Decision tree는 information based machine learning
- 그러므로 쉐넌이 가장 먼저 도입했던 정보에 대한 공식을 기반으로 함. 
- 정보를 수량화하기 위해 나온 식이며, 정보는 degree of surprisal로 사건이 일어날 가능성이 클수록 정보량이 작다.
- 즉 역의 관계를 보임. (ex)내일 해가 뜬다 -> 당연한 정보, 정보량이 작음)
- Entropy는 정보량(log2(P(x)))의 기댓값(평균)
- 엔트로피가 크다 = 정보량이 크다. 정보량이 많다. 즉, 우리가 그 만큼 질문(descriptive feature)을 던지는 갯수가 많이 필요하다. *Decision tree와 Entropy*
- 1/확률 -> 확률이 커질수록 정보량 작아짐. 

#### Information Gain
- 모집단에서 Descriptive feature 기준으로 분기함.
- 분기 전의 Entropy와 분기 후의 총 Entropy(가중치/weight 고려) 차이 = Information Gain

#### Decision Tree 그리기

- Vegetation classification dataset으로 Decision Tree 그리기.
1. 전체의 Entropy 계산
2. 각 descriptive feature 별로 Entropy 계산 후 IG값이 가장 큰 것으로 분기.

```python
#전체 엔트로피
H=-(3/7*np.log2(3/7))-2/7*np.log2(2/7)*2
print(H)
```

```python
#Stream
Hsf=-(2/3*np.log2(2/3))-1/3*np.log2(1/3)
Hst=-(1/2*np.log2(1/2))-1/4*np.log2(1/4)-1/4*np.log2(1/4)
IGS=H-3/7*Hsf-4/7*Hst
print(IGS)
```

```python
#Slope
Hs=-(3/5*np.log2(3/5))-1/5*np.log2(1/5)-1/5*np.log2(1/5)
Hm=0
Hf=0
IG=H-Hs*5/7-Hm-Hf
print(IG)
```

```python
#Elevation
Heh=-(2/3*np.log2(2/3))-1/3*np.log2(1/3)
Hel=0
Hem=1
Het=0
IGE=H-3/7*Heh-1/7*Hel-2/7*Hem-1/7*Het
print(IGE)
```

- 계산 결과 Elevation의 Inforamtion gain이 가장 큼.
- Elevation을 root node로 사용하여 Decision tree 그리기.
